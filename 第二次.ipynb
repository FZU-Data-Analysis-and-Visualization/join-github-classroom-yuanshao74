```python
import numpy as np
arr = np.arange(1, 13)
arr_2d = arr.reshape(3, 4)
print("数组的维度 (ndim):", arr_2d.ndim)
print("数组的形状 (shape):", arr_2d.shape)
print("数组的元素总个数 (size):", arr_2d.size)
```

```python
import numpy as np
arr_float=np.array([1.5, 2.6, 3.7, 4.8])
arr_int = arr_float.astype(np.int32)
print("转换前的数组:", arr_float)
print("转换前的数据类型:", arr_float.dtype.name)
print("转换后的数组:", arr_int)
print("转换后的数据类型:", arr_int.dtype.name)

      
```

```python
import numpy as np
arr = np.array([10, 15, 20, 25, 30, 35])
fancy_result = arr[[0, 2, 4]]
bool_result = arr[arr > 20]
print("1. 花式索引结果:", fancy_result)
print("2. 布尔索引结果:", bool_result)
```

```python
import numpy as np
arr_zeros = np.zeros((3, 4))
arr_b = np.array([1, 2, 3, 4])
result = arr_zeros + arr_b
print("全0数组:\n", arr_zeros)
print("\n一维数组:", arr_b)
print("\n相加后的结果:\n", result)
```

六

```python
import numpy as np
x = np.array([1, 4, 9, 16])
y = np.array([2, 1, 4, 5])
sqrt_result = np.sqrt(x)
max_result = np.maximum(sqrt_result, y)
print("1. 平方根数组:", sqrt_result)
print("2. 对应位置的最大值:", max_result)
```

```python
import numpy as np
arr_x = np.array([1, 5, 7])
arr_y = np.array([2, 6, 8])
arr_con = np.array([True, False, True])
result = np.where(arr_con, arr_x, arr_y)
print("条件逻辑运算结果:", result)

```

```python
import numpy as np
arr = np.array([[6, 2, 7], 
                [3, 6, 2], 
                [4, 3, 2]])
mean_value = np.mean(arr)
print("1. 数组所有元素的算术平均值:", mean_value)
sorted_arr = np.sort(arr, axis=0) 
print("2. 按列排序后的结果:")
print(sorted_arr)
```
