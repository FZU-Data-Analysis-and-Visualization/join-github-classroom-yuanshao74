```python

df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

res1 = df[["score", "gender"]]
print("选取两列：\n", res1)

res2 = df.iloc[[0, 2]]
print("\n选取索引0、2行：\n", res2)

res3 = df[df["score"] > 80]
print("\nscore>80：\n", res3)

df["grade"] = df["score"].apply(lambda x: "优秀" if x >= 90 else "良好")
print("\n新增grade列：\n", df)

df = df.drop("gender", axis=1)
print("\n删除gender列：\n", df)
```

```python

df_null = pd.DataFrame({
    "x": [1, None, 3],
    "y": [None, 5, 6]
})
print("缺失值位置：\n", df_null.isnull())
print("\n每列缺失值个数：\n", df_null.isnull().sum())

df_fill = df_null.fillna(0)
print("\n0填充缺失值：\n", df_fill)

df_drop = df_null.dropna()
print("\n删除缺失行：\n", df_drop)
```

```python
df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

res1 = df.groupby("gender")["score"].agg(["mean", "max"])
print("分组均值、最大值：\n", res1)

res2 = df.groupby("gender").size()
print("\n每组人数：\n", res2)
```

```python
df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

df_sort = df.sort_values("score")
print("按score升序：\n", df_sort)
df = pd.concat([df, pd.DataFrame({"name":["Tom"],"score":[85],"gender":["男"]})], ignore_index=True)
print("\n添加重复行后：\n", df)

df_dup = df.drop_duplicates()
print("\n去重后：\n", df_dup)
```

```python

df1 = pd.DataFrame({"id":[1,2], "name":["X","Y"]})
df2 = pd.DataFrame({"id":[1,2], "age":[20,30]})
df_merge = pd.merge(df1, df2, on="id")
print("横向合并：\n", df_merge)

df_a = pd.DataFrame({"val":[1,2]})
df_b = pd.DataFrame({"val":[3,4]})
df_concat = pd.concat([df_a, df_b], ignore_index=True)
print("\n纵向拼接：\n", df_concat)
```

```python

date_list = ["2025-01-01","2025-01-02","2025-01-03"]
df_time = pd.DataFrame({"date": date_list})
df_time["date"] = pd.to_datetime(df_time["date"])
print("转为时间格式：\n", df_time)

df_time["month"] = df_time["date"].dt.month
df_time["weekday"] = df_time["date"].dt.day_name()
print("\n提取月份、星期：\n", df_time)
```

```python
df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

df["score_norm"] = (df["score"] - df["score"].mean()) / df["score"].std()
print("标准化后：\n", df)

pivot_df = df.pivot_table(index="gender", values="score", aggfunc="median")
print("\n透视表（中位数）：\n", pivot_df)
```
